```python
# create a mapping dictionary and than map all the deseases with numarical values (lable encoding)

mapDict = {}
for index, element in enumerate(disease):
    mapDict[index] = element
mapDict = dict(map(reversed, mapDict.items()))

data_train.replace(mapDict,inplace=True)
data_test.replace(mapDict,inplace=True)
    


#check the data_train 
#print(data_train.head())

X= data_train[l1]

#print(X)

y = data_train[["prognosis"]]
np.ravel(y)




X_test= data_test[l1]
y_test = data_test[["prognosis"]]

#print(y_test)

np.ravel(y_test)
```

```python
# defining function Decision Tree

def DecisionTree():

    from sklearn import tree

    dt = tree.DecisionTreeClassifier() 
    dt = dt.fit(X,y)

    from sklearn.metrics import accuracy_score
    y_pred=dt.predict(X_test)
    print(accuracy_score(y_test, y_pred))

    psymptoms = [Symptom1.get(),Symptom2.get(),Symptom3.get(),Symptom4.get(),Symptom5.get()]

    for k in range(len(l1)):
        for z in psymptoms:
            if(z==l1[k]):
                l2[k]=1

    inputtest = [l2]
    predict = dt.predict(inputtest)
    predicted=predict[0]

    h='no'
    for a in range(len(disease)):
        if(predicted == a):
            h='yes'
            break


    if (h=='yes'):
        t1.delete("1.0", END)
        t1.insert(END, disease[a])
    else:
        t1.delete("1.0", END)
        t1.insert(END, "Not Found")

```

```python
# defining function Random Forest


def randomforest():
    from sklearn.ensemble import RandomForestClassifier
    rf1 = RandomForestClassifier()
    rf1 = rf1.fit(X,np.ravel(y))

    # calculating accuracy 
    from sklearn.metrics import accuracy_score
    y_pred=rf1.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    
    psymptoms = [Symptom1.get(),Symptom2.get(),Symptom3.get(),Symptom4.get(),Symptom5.get()]

    for k in range(0,len(l1)):
        for z in psymptoms:
            if(z==l1[k]):
                l2[k]=1

    inputtest = [l2]
    predict = rf1.predict(inputtest)
    predicted=predict[0]

    h='no'
    for a in range(len(disease)):
        if(predicted == a):
            h='yes'
            break

    if (h=='yes'):
        t2.delete("1.0", END)
        t2.insert(END, disease[a])
    else:
        t2.delete("1.0", END)
        t2.insert(END, "Not Found")

```

```python
# defining function NaiveBayes


def NaiveBayes():
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb=gnb.fit(X,np.ravel(y))

    from sklearn.metrics import accuracy_score
    y_pred=gnb.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    
    psymptoms = [Symptom1.get(),Symptom2.get(),Symptom3.get(),Symptom4.get(),Symptom5.get()]
    for k in range(len(l1)):
        for z in psymptoms:
            if(z==l1[k]):
                l2[k]=1

    inputtest = [l2]
    predict = gnb.predict(inputtest)
    predicted=predict[0]

    h='no'
    for a in range(len(disease)):
        if(predicted == a):
            h='yes'
            break

    if (h=='yes'):
        t3.delete("1.0", END)
        t3.insert(END, disease[a])
    else:
        t3.delete("1.0", END)
        t3.insert(END, "Not Found")
        
        
     
```

```python
# defining function Support Vector Machine


from sklearn.svm import SVC
def SupportVectorMachine():
    svm = SVC(kernel='linear')
    svm.fit(X, np.ravel(y))

    from sklearn.metrics import accuracy_score
    y_pred = svm.predict(X_test)
    print(accuracy_score(y_test, y_pred))

    psymptoms = [Symptom1.get(), Symptom2.get(), Symptom3.get(), Symptom4.get(), Symptom5.get()]

    for k in range(0, len(l1)):
        for z in psymptoms:
            if z == l1[k]:
                l2[k] = 1

    inputtest = [l2]
    predict = svm.predict(inputtest)
    predicted = predict[0]

    h = 'no'
    for a in range(len(disease)):
        if predicted == a:
            h = 'yes'
            break

    if h == 'yes':
        t4.delete("1.0", END)
        t4.insert(END, disease[a])
    else:
        t4.delete("1.0", END)
        t4.insert(END, "Not Found")
```
